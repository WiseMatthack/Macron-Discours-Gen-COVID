Nous pouvons désormais générer notre discours ! Vous pouvez choisir le nombre de lignes souhaitées en changeant la valeur de "nb_lignes".

```
#@title Générateur de discours
 
import requests
import markovify
import random
 
# settings
nb_lignes =  30#@param {type: "number"}
 
#getting script
discours_url = 'https://raw.githubusercontent.com/WiseMatthack/Macron-Discours-Gen-COVID/master/macron-discours.txt'
discours_request = requests.get(discours_url, allow_redirects=True)
 
open('macron-discours.txt', 'wb').write(discours_request.content)
 
# Get raw text as string.
with open("macron-discours.txt") as f:
    discours = f.read()
 
# Build the model.
discours_model = markovify.Text(discours)
 
for i in range(nb_lignes):
    print(discours_model.make_sentence())
```
